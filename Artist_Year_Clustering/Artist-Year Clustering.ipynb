```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Description of dataset

Each column represents, for each year, the fraction of times an artist has been listened to. Data comes from my own LastFM account, containing data from 2014 to 2018.

```python
df = pd.read_csv('data.csv', sep=';')
# Anonimizing data
df.Artist = ['Artist{}'.format(i) for i in range(len(df))]
df.head(10)
```

```python
df.describe()
```

### Considering only last two years

```python
df[['Artist', '2017', '2018']].head(10)
```

```python
plt.scatter(df['2017'], df['2018'])
```

```python
X = df.values[:,4:6]
X.shape
```

#### Experiment with KMeans

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

fig, ax = plt.subplots(3, 2, figsize=(10, 15))
for i, n_clusters in enumerate(range(2, 8)):
    km = KMeans(n_clusters=n_clusters)
    km.fit(X)
    y = km.predict(X)
    df['cluster'] = y
    cur_ax = ax[i//2, i%2]
    sns.scatterplot(x='2017', y='2018', data=df, hue='cluster', 
                    palette=sns.mpl_palette('tab10', np.max(df.cluster+1)), ax=cur_ax, legend = False)
    cur_ax.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], marker='x', color='k')
    cur_ax.set_title('{} Clusters. Silhouette Score: {:.3f}'.format(n_clusters, silhouette_score(X, y)))
```

#### Experiment with Agglomerative Clustering and custom distance metric

Distance metric:

```python
dist_metric = lambda x, y: np.sum((x-y)**2)
```

```python
from sklearn.cluster import AgglomerativeClustering

fig, ax = plt.subplots(3, 2, figsize=(10, 15))
for i, n_clusters in enumerate(range(2, 8)):
    agglo = AgglomerativeClustering(n_clusters=n_clusters)
    y = agglo.fit_predict(X)
    df['cluster'] = y
    cur_ax = ax[i//2, i%2]
    sns.scatterplot(x='2017', y='2018', data=df, hue='cluster', 
                    palette=sns.mpl_palette('tab10', np.max(df.cluster+1)), ax=cur_ax, legend = False)
    cur_ax.set_title('{} Clusters'.format(n_clusters))
```

```python
agglo = AgglomerativeClustering(n_clusters=6)
y = agglo.fit_predict(X)
df['cluster'] = y

df.sample(20)
```

```python
sns.scatterplot(x='2017', y='2018', data=df, hue='cluster', 
                    palette=sns.mpl_palette('tab10', np.max(df.cluster+1)))
```

```python
def plot_clustering(df):
    fig_height = 0.3*len(df)
    plt.figure(figsize=(9, fig_height))
    accum = np.zeros(len(df))
    for col in df.columns[1:-1]:
        plt.barh(y=df.Artist+' - '+df.cluster.map(str), width=df[col], left=accum, label=col)
        accum += df[col]
    plt.xlabel('Proportion')
    plt.ylabel('Artist and Cluster')
    plt.legend()
```

```python
df_temp = df[['Artist', '2017', '2018', 'cluster']].sample(30)
plot_clustering(df_temp[['Artist', '2017', '2018', 'cluster']].sort_values(by=['cluster', '2017', '2018']))
```

### Using 5 Dimensions

```python
X = df.values[:,1:6]
X.shape
```

```python
agglo = AgglomerativeClustering(n_clusters=8)
y = agglo.fit_predict(X)
df['cluster'] = y

df.sample(20)
```

```python
df_temp = df
plot_clustering(df_temp.sort_values(by=['cluster', '2017', '2018']))
plt.savefig('clusters.png')
```

```python
df_grouped = df.groupby('cluster').mean()

fig_height = 0.3*len(df_grouped)
plt.figure(figsize=(9, fig_height))

accum = np.zeros(len(df_grouped))
for col in df_grouped.columns:
    plt.barh(y=df_grouped.index.map(str), width=df_grouped[col], left=accum, label=col)
    accum += df_grouped[col]
plt.xlabel('Proportion')
plt.ylabel('Cluster')
plt.legend()
df_grouped
```
